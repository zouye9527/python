1.创建字典

```python
fruit={
    "apple":"苹果",
    "banana":"香蕉",
    "orange":"橘子"
}
print(fruit)
```

2.通过键找值

```python
print(fruit['apple'])
print(fruit['orange'])
```

3.通过键修改值

```python
fruit['banana']='香甜的香蕉'
print(fruit['banana'])
```

4.添加键值对：直接给不存在的键添加值，就可以

```python
fruit['grape']='葡萄'
fruit[4]='序号4'
print(fruit)
```

5.获取字典所有的键

```python
print(fruit.keys())
```

6.获取字典所有的值

```python
print(fruit.values())
```

7.检查某个键是否在字典中

```python
print('apple' in fruit)
print('pear' in fruit)
```

8.获取字典的长度：返回键值对的个数

```python
print(len(fruit))
```

9.使用get()方法访问值，可以避免报错

```python
print(fruit.get('apple','字典中没有这个键值'))
print(fruit.get('peach','字典中没有这个键值对'))
```

10.遍历字典1：按键遍历

```python
for key in fruit:
    print(f'键：{key}，值：{fruit[key]}')
```

11.遍历字典2：直接遍历键值对,使用item()方法

```python
for key,value in fruit.items():
    print(f'键：{key}，值：{value}')
```

12.字典的修改。如果有键，更新值；否则直接添加。update()方法

```python
fruit.update({
    'banana':'美味香甜的香蕉',
    6:'序号6'
})
print(fruit)
```

13.字典的清空。clear()方法

```python
fruit.clear()
print(fruit)
```
