1.创建字典

```python
fruit={
    "apple":"苹果",
    "banana":"香蕉",
    "orange":"橘子"
}
print(fruit)
```

2.通过键找值

```python
print(fruit['apple'])
print(fruit['orange'])
```

3.通过键修改值

```python
fruit['banana']='香甜的香蕉'
print(fruit['banana'])
```

4.添加键值对：直接给不存在的键添加值，就可以

```python
fruit['grape']='葡萄'
fruit[4]='序号4'
print(fruit)
```

5.获取字典所有的键

```python
print(fruit.keys())
```

6.获取字典所有的值

```python
print(fruit.values())
```

7.检查某个键是否在字典中

```python
print('apple' in fruit)
print('pear' in fruit)
```

8.获取字典的长度：返回键值对的个数

```python
print(len(fruit))
```
