1.创建集合。集合中不能有重复的元素

```python
s1={1,2,3,3,'python'}
print(s1)
```

2.把列表转换为集合。自动去重

```python
l1=[1,2,3,3]
s2=set(l1)
print(s2)
```

3.把字符串转换为集合。去重，顺序会变。

```python
st='hello'
s3=set(st)
print(s3)
```

4.空集合只能用set创建

```python
s4=set()
s5={}
print(type(s4))
print(type(s5))
```

5.增加元素

```python
s6={1,2,3,4}
s6.add(2)
print(s6)
s6.add(5)
print(s6)
```
